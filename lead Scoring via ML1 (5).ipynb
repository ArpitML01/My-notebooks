# 6.a. Check for multicollinearity through VIF (variance inflation factor)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# Make a copy to avoid modifying the original dataframe
df_vif = df1.copy()

# Drop categorical and non-numeric columns
df_vif = df_vif.select_dtypes(include=[np.number])

# Drop the target variable 'Converted' as it's not an independent variable
if 'Converted' in df_vif.columns:
    df_vif = df_vif.drop(columns=['Converted'])

# Standardize numerical data for better VIF calculation
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_vif), columns=df_vif.columns)

# Compute VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = df_vif.columns
vif_data["VIF"] = [variance_inflation_factor(df_scaled.values, i) for i in range(df_scaled.shape[1])]

# Sort VIF values in descending order
vif_data = vif_data.sort_values(by="VIF", ascending=False)

# Display VIF values
print(vif_data)
```

#6.b. Check for non-correlated variables through Heat matrix or also through Code for high correlation matrix

```python
# Assuming df1 is the modified dataframe after dropping some columns from df
df1_columns = df1.columns  # Get columns present in df1

# Compute the correlation matrix only for df1 columns
corr_matrix_df1 = df1[df1_columns].corr(numeric_only=True)

# Plot heatmap for df1
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix_df1, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Filtered Columns)")
plt.show()

# Find highly correlated variables (threshold = 0.8)
threshold = 0.8
high_corr_pairs_df1 = []

for i in range(len(corr_matrix_df1.columns)):
    for j in range(i):
        if abs(corr_matrix_df1.iloc[i, j]) > threshold:
            high_corr_pairs_df1.append((corr_matrix_df1.columns[i], corr_matrix_df1.columns[j], corr_matrix_df1.iloc[i, j]))

# Display high correlation pairs
if high_corr_pairs_df1:
    print("\nHighly Correlated Variables (|r| > 0.8) in df1:")
    for var1, var2, corr_value in high_corr_pairs_df1:
        print(f"{var1} & {var2} -> Correlation: {corr_value:.2f}")
else:
    print("No highly correlated variables found in df1.")
```

#  Feature ranking through REcursive feature elimination (RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression  # Change model if needed

# Define target (y) and features (X)
X = df1.drop("Converted", axis=1)  # Replace 'Target_Column' with your actual target column
y = df1["Converted"]

# Initialize the model (use Logistic Regression or other estimators)
model = LogisticRegression(max_iter=1000)

# Define RFE (selecting the top 'n' features)
n_features_to_select = 10  # Change as per requirement
rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)

# Fit RFE
rfe.fit(X, y)

# Print selected features
selected_features = X.columns[rfe.support_]
print("\nSelected Features:")
print(selected_features)

# Print feature ranking
feature_ranking = dict(zip(X.columns, rfe.ranking_))
print("\nFeature Ranking (Lower is better):")
print(feature_ranking)
```
